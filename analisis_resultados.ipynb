```python
import pandas as pd
import numpy as np

# Leer el archivo CSV
path = './dataset.csv'
data = pd.read_csv(path, delimiter=';', index_col=False)

# Excluir 'Acebeda (La)'
data = data[data['Nombre'] != 'Acebeda (La)']

# Definir grupos de edad activos y nombre de la ratio
grupos_edad_activos = ['poblacion_censada_{}a{}'.format(i, i+4) for i in range(15, 65, 5)]
ratio_calc = 'ratio_contratos'

# Función para calcular estadísticas por cluster
def calcular_estadisticas(data, cluster_col):
    estadisticas = {}
    for edad in grupos_edad_activos:
        col_name = f'{ratio_calc}_por_{edad}'
        stats = data.groupby(cluster_col)[col_name].agg(['std']).reset_index()
        cov = data.groupby(cluster_col)[col_name].apply(lambda x: np.cov(x)).reset_index(name='cov')
        merged_stats = pd.merge(stats, cov, on=cluster_col)
        estadisticas[edad] = merged_stats
    return estadisticas

# Calcular estadísticas para cada cluster
estadisticas_clusters = {}
for cluster_col in ['cluster_poblacion_censada', 'cluster_distancia_capital', 'cluster_zona_estadistica', 'cluster_densidad_poblacion']:
    cluster_name = cluster_col.split('_')[1]
    estadisticas_clusters[cluster_name] = calcular_estadisticas(data, cluster_col)

# Mostrar las estadísticas
for cluster_name, estadisticas in estadisticas_clusters.items():
    print(f'\nEstadísticas para {cluster_name}:')
    for edad, stats in estadisticas.items():
        print(f'\n{edad}:\n', stats)
```

```python
import matplotlib.pyplot as plt

# Colores para los clusters
cluster_colors = {
    0: '#B3E5FC',
    1: '#DCEDC8',
    2: '#FFCCBC',
}

for cluster_col in ['cluster_poblacion_censada', 'cluster_distancia_capital', 'cluster_zona_estadistica', 'cluster_densidad_poblacion']:
    cluster_name = cluster_col.split('_')[1]
    estadisticas = estadisticas_clusters[cluster_name]

    fig, axs = plt.subplots(1, len(grupos_edad_activos), figsize=(18, 4))
    for i, edad in enumerate(grupos_edad_activos):
        stats = estadisticas[edad]
        axs[i].bar(stats[cluster_col], stats['std'], color=[cluster_colors[c] for c in stats[cluster_col]])
        axs[i].set_title(f'{edad}')
        axs[i].set_xlabel(cluster_name)
        axs[i].set_ylabel('Desviación Típica')

    plt.suptitle(f'Desviación Típica del {ratio_calc} por {cluster_name}')
    plt.tight_layout(rect=[0, 0, 1, 0.95])
    plt.show()
```

```python
for cluster_col in ['cluster_poblacion_censada', 'cluster_distancia_capital', 'cluster_zona_estadistica', 'cluster_densidad_poblacion']:
    cluster_name = cluster_col.split('_')[1]
    estadisticas = estadisticas_clusters[cluster_name]

    fig, axs = plt.subplots(1, len(grupos_edad_activos), figsize=(18, 4))
    for i, edad in enumerate(grupos_edad_activos):
        stats = estadisticas[edad]
        axs[i].bar(stats[cluster_col], stats['cov'], color=[cluster_colors[c] for c in stats[cluster_col]])
        axs[i].set_title(f'{edad}')
        axs[i].set_xlabel(cluster_name)
        axs[i].set_ylabel('Covarianza')

    plt.suptitle(f'Covarianza del {ratio_calc} por {cluster_name}')
    plt.tight_layout(rect=[0, 0, 1, 0.95])
    plt.show()
```
